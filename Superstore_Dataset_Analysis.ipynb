#### Q5. Which product Category appears the most number of times?

```python
category_counts = df['Category'].value_counts()

most_category = category_counts.idxmax()
count = category_counts.max()

print("Most Frequent Category:", most_category)
print("Number of Occurrences:", count)
```

##### Answer

The product category that appears the most number of times is **Office Supplies**.

It appears **6026 times** in the dataset.
